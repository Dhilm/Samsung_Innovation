```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("school_kzinfo.csv")
```

```python
age_groups = ["3-6", "7-10", "11-15", "16-17"]
filtered_data = data[data['Age Group'].isin(age_groups)]
```

```python
population_totals = filtered_data.groupby('Age Group')['Population'].sum()

plt.figure(figsize=(8, 8))
plt.pie(population_totals, labels=population_totals.index, autopct='%1.1f%%', startangle=140)
plt.title('Population Distribution by Age Group')
plt.axis('equal') 
plt.show()
```

```python
age_group = "7-10"
age_group_data = data[data['Age Group'] == age_group]

plt.figure(figsize=(10, 6))
plt.bar(age_group_data['Year'], age_group_data['Population'])
plt.title(f'Population Trend for Age Group {age_group}')
plt.xlabel('Year')
plt.ylabel('Population')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
