```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("country_static.csv")
```

```python
plt.figure(figsize=(8, 8))
plt.pie(data['Expenditure'], labels=data['Country'], autopct='%1.1f%%')
plt.title('Education Expenditure by Country')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(data['Country'], data['Expenditure'], color='skyblue')
plt.xlabel('Country')
plt.ylabel('Expenditure')
plt.title('Education Expenditure by Country')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['Country'], data['Expenditure'], color='orange')
plt.xlabel('Country')
plt.ylabel('Expenditure')
plt.title('Education Expenditure by Country')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data['Country'], data['Expenditure'], marker='o', color='green', linestyle='-', linewidth=2)
plt.xlabel('Country')
plt.ylabel('Expenditure')
plt.title('Education Expenditure by Country')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
import folium

country_coordinates = {
    "Russia": [61.524, 105.3188],
    "Kazakhstan": [48.0196, 66.9237],
    "Ukraine": [48.3794, 31.1656],
    "Belarus": [53.7098, 27.9534],
    "Uzbekistan": [41.3775, 64.5853],
    "Turkmenistan": [38.9697, 59.5563],
    "Germany": [51.1657, 10.4515],
    "France": [46.6034, 1.8883],
    "United Kingdom": [55.3781, -3.4360],
    "Italy": [41.8719, 12.5675],
    "Spain": [40.4637, -3.7492],
    "Netherlands": [52.1326, 5.2913]
}

m = folium.Map(location=[51.5074, 0.1278], zoom_start=2)

for i in range(len(data)):
    country = data['Country'][i]
    expenditure = data['Expenditure'][i]
    
    coordinates = country_coordinates.get(country)
    
    if coordinates is not None:
        folium.Marker(
            location=coordinates,
            popup=f"{country}: {expenditure}",
            tooltip=country,
        ).add_to(m)


m
```
