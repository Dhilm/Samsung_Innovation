```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("dead_info.csv")
```

```python
print(data.head())
```

```python
plt.figure(figsize=(8, 8))
data['Cause of Death'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Causes of Death')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
grouped_data = data.groupby(['Country', 'Cause of Death'])['Count'].sum().reset_index()
for cause in grouped_data['Cause of Death'].unique():
    subset = grouped_data[grouped_data['Cause of Death'] == cause]
    plt.bar(subset['Country'], subset['Count'], label=cause)

plt.title('Number of Deaths by Country and Cause')
plt.xlabel('Country')
plt.ylabel('Number of Deaths')
plt.xticks(rotation=45)
plt.legend(title='Cause of Death')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
for cause in data['Cause of Death'].unique():
    subset = data[data['Cause of Death'] == cause]
    plt.scatter(subset['Country'], subset['Count'], label=cause)

plt.title('Number of Deaths by Country and Cause')
plt.xlabel('Country')
plt.ylabel('Number of Deaths')
plt.xticks(rotation=45)
plt.legend(title='Cause of Death')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
for cause in data['Cause of Death'].unique():
    subset = data[data['Cause of Death'] == cause]
    plt.plot(subset['Country'], subset['Count'], marker='o', label=cause)

plt.title('Number of Deaths by Country and Cause')
plt.xlabel('Country')
plt.ylabel('Number of Deaths')
plt.xticks(rotation=45)
plt.legend(title='Cause of Death')
plt.tight_layout()
plt.show()
```

```python
country_coordinates = {
    'USA': (37.0902, -95.7129),
    'Kazakhstan': (48.0196, 66.9237),
    'Russia': (61.5240, 105.3188),
    'Africa': (-8.7832, 34.5085)
}
plt.figure(figsize=(12, 8))

for country, count in zip(data['Country'], data['Count']):
    if country in country_coordinates:
        plt.plot(country_coordinates[country][1], country_coordinates[country][0], 'o', markersize=count/5000, label=country)

plt.title('Number of Deaths by Country')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend()
plt.grid(True)
plt.show()
```
